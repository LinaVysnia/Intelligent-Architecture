```python
from huggingface_hub import login
from datasets import load_dataset_builder,load_dataset
import matplotlib.pyplot as plt
import pandas as pd
```

```python
login()
ds_builder = load_dataset_builder("zimhe/pseudo-floor-plan-12k")
ds_builder.info.description
ds_builder.info.features
```

```python
ds_plans = load_dataset("zimhe/pseudo-floor-plan-12k")
ds_plans.cache_files
ds_plans.num_columns
print(ds_plans)
```

```python
sample = ds_plans['train'][50]
#indice = sample['indices']  
plan = sample['plans']  
wall = sample['walls'] 
color = sample['colors']  
footprint = sample['footprints'] 
#captions = sample['captions'] 

fig, axes = plt.subplots(2, 2, figsize=(10, 10))
axes[0, 0].imshow(plan)
axes[0, 0].set_title('Plan')
axes[0, 1].imshow(wall)
axes[0, 1].set_title('Walls')
axes[1, 0].imshow(color)
axes[1, 0].set_title('Colors')
axes[1, 1].imshow(footprint)
axes[1, 1].set_title('Footprints')

plt.tight_layout()
plt.show()
```
